```python
# Average Monthly Charges by Contract Type — compare typical spending per contract.
Average_Monthly_Charges_by_Contract_Type = df.groupby('Contract')['MonthlyCharges'].mean().reset_index()
Average_Monthly_Charges_by_Contract_Type
```

```python
sns.barplot(data=Average_Monthly_Charges_by_Contract_Type, x='Contract', y='MonthlyCharges', palette='viridis')
plt.title('Average Monthly Charges by Contract Type')
plt.show()
```

```python
# Correlation Heatmap — between numerical columns (e.g., tenure, charges, total charges).
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
plt.figure(figsize = (8,6))
sns.heatmap(corr_matrix,annot=True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Corelation between Numerical Values')
plt.show()
```

```python
# Pie chart: Payment Method Distribution
payment_counts = df['PaymentMethod'].value_counts().plot(kind = 'pie', autopct='%1.1f%%')
plt.title('Payment Method Distribution')
plt.ylabel('')  # hides the y-label for a cleaner look
plt.show()
```

```python
# Bar Plot: Churn Count by Payment Method
Churn_Count_by_Payment_Method = df.groupby('Churn')['PaymentMethod'].count().reset_index()
Churn_Count_by_Payment_Method
```

```python
sns.barplot(data = Churn_Count_by_Payment_Method, x = 'Churn', y = 'PaymentMethod' ,palette='viridis')
plt.title('Churn Count by Payment Method')
plt.show()
```

#### ## 🧠 Project Summary: Telco Customer Churn Analysis

### Objective
The goal of this project was to analyze customer data from a telecom company to understand patterns in customer churn and identify key factors that influence churn.

### Key Steps
- **Data Cleaning:** Handled missing values, converted data types, and categorized tenure.
- **Feature Engineering:** Created new columns like Tenure_Group for better segmentation.
- **EDA:** Explored relationships between churn and contract type, gender, internet service, and payment method.
- **Visualization:** Used Seaborn and Matplotlib to create bar charts, pie charts, and heatmaps.

### Key Insights
- Customers on **Month-to-Month contracts** have the highest churn rate.
- **Fiber optic** users churn more than DSL users.
- **Electronic check** payment users have higher churn rates compared to credit card or bank transfer users.
- Customers with **longer tenure** or **long-term contracts** are less likely to churn.

### Conclusion
Retention strategies should target month-to-month and fiber optic customers. Encouraging longer contracts and offering loyalty benefits can help reduce churn.
